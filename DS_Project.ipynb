```python
def return_PE(name):
    name1 = name +".BK"
    stock = yf.Ticker(name1)
    print(name1)
    info = stock.info
    # Check if 'trailingEps' key exists in the dictionary
    if "trailingEps" not in info or "currentPrice" not in info:
        PE = 0
        return PE

    # Check for None or NaN values
    if info["trailingEps"] is None or info["currentPrice"] is None or info["trailingEps"] == 'NaN' or info["currentPrice"] == 'NaN':
        PE = 0
        return PE

    price = info["currentPrice"]
    EPS = info["trailingEps"]
    PE = round(price / EPS, 2)
    print(PE)
    return PE
```

```python
def return_PBV(name):
    name1 = name +".BK"
    stock = yf.Ticker(name1)
    print(name1)
    info = stock.info

    # Check if 'bookValue' key exists in the dictionary
    if "bookValue" not in info or "currentPrice" not in info:
        P_BV = 0
        return P_BV

    # Check for None or NaN values
    if info["bookValue"] is None  or info["bookValue"] == 'NaN' or info["currentPrice"] is None  or info["currentPrice"] == 'NaN' :
        P_BV = 0
        return P_BV


    BV = info["bookValue"]
    price = info["currentPrice"]
    P_BV = round(price / BV, 2)
    print (P_BV)

    return P_BV
```

```python
def return_cashflow(name):
    name1 = name +".BK"
    stock = yf.Ticker(name1)
    print(name1)
    info = stock.info

    # Check if 'bookValue' key exists in the dictionary
    if "freeCashflow" not in info :
        cashflow = 0
        return cashflow

    # Check for None or NaN values
    if info["freeCashflow"] is None or info["freeCashflow"] == 'NaN' :
        cashflow = 0
        return cashflow

    Cashflow = info["freeCashflow"]
    print(Cashflow)
    return Cashflow
```

```python
def return_payoutRatio(name):
    name1 = name +".BK"
    stock = yf.Ticker(name1)
    print(name1)
    info = stock.info

    # Check if 'payoutRatio' key exists in the dictionary
    if "payoutRatio" not in info :
        Ratio = 0
        return Ratio

    # Check for None or NaN values
    if info["payoutRatio"] is None or info["payoutRatio"] == 'NaN' :
        Ratio = 0
        return Ratio

    Ratio = info["payoutRatio"]
    print(Ratio)
    return Ratio
```

```python
for index, row in set100.iterrows():
    set100.loc[index, 'P/E'] = return_PE(row['Symbol'])
```

```python
for index, row in set100.iterrows():
    set100.loc[index, 'P/BV'] = return_PBV(row['Symbol'])
```

```python
for index, row in set100.iterrows():
    set100.loc[index, 'Cash Flow'] = return_cashflow(row['Symbol'])
```

```python
for index, row in set100.iterrows():
    set100.loc[index, 'Payout Ratio'] = return_payoutRatio(row['Symbol'])
```
